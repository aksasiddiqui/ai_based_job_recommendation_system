```python
import pandas as pd
```

```python
Jobs = pd.read_csv("raw_data/postings.csv")
Jobs.shape
```

```python
#shuffling the dataset to remove any sort of bias

Jobs_shuffled = Jobs.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
#first 10000 rows taken 
FinalJobs = Jobs_shuffled.iloc[:10000]
```

```python
FinalJobs.columns, FinalJobs.shape
```

```python
FinalJobs.sample(1)
```

```python
#  Create "working_mode" column from "Work Type"


FinalJobs["working_mode"] = FinalJobs["work_type"]

FinalJobs["salary"] = FinalJobs["normalized_salary"]

FinalJobs["position"] = FinalJobs["title"]

FinalJobs["job_role_and_duties"] = (
    FinalJobs["title"].fillna("") + " " +
    FinalJobs["description"].fillna("")
)

FinalJobs["requisite_skill"] = FinalJobs["skills_desc"]

# Drop unnecessary columns (optional)
FinalJobs = FinalJobs.drop(columns=[
    "max_salary",
    "med_salary",
    "normalized_salary",
    "compensation_type",

    "views",
    "sponsored",
    "posting_domain",
    "zip_code",
    "fips",

    "title",
    "description",
    "skills_desc",
    "work_type"
], errors="ignore")
```

```python
FinalJobs.columns
```

```python
FinalJobs = FinalJobs.drop(columns=[
    "min_salary",
    "formatted_work_type",
    "applies",
    "application_url",
    "application_type",
    "closed_time",
    "formatted_experience_level"
], errors="ignore")
```

```python
FinalJobs.columns
```

```python
FinalJobs = FinalJobs.drop(columns=[
    "pay_period",
    "original_listed_time",
    "job_posting_url",
    "expiry"
], errors="ignore")
```

```python
text_cols = ["requisite_skill", "job_role_and_duties", "position"]
for col in text_cols:
    FinalJobs[col].fillna("Not Provided", inplace=True)

cat_cols = ["remote_allowed", "working_mode", "currency"]
for col in cat_cols:
    FinalJobs[col].fillna("Unknown", inplace=True)
```

```python
for i in FinalJobs.columns:
  print( i , "is   ", FinalJobs[i][0])  
```

```python
import pandas as pd
import numpy as np

# Make copy to avoid warnings
FinalJobs = FinalJobs.copy()

# Replace -1 with NaN first
FinalJobs["salary"].replace(-1, np.nan, inplace=True)

# Convert to numeric
FinalJobs["salary"] = pd.to_numeric(FinalJobs["salary"], errors="coerce")

# Fill NaN with mean salary
mean_salary = FinalJobs["salary"].mean()
FinalJobs["salary"].fillna(mean_salary, inplace=True)

```

```python
# Convert text columns to lowercase and strip extra spaces
FinalJobs["location"] = FinalJobs["location"].astype(str).str.lower().str.strip()
FinalJobs["working_mode"] = FinalJobs["working_mode"].astype(str).str.lower().str.strip()
FinalJobs["position"] = FinalJobs["position"].astype(str).str.lower().str.strip()
FinalJobs["job_role_and_duties"] = FinalJobs["job_role_and_duties"].astype(str).str.lower().str.strip()
FinalJobs["requisite_skill"] = FinalJobs["requisite_skill"].astype(str).str.lower().str.strip()
```

```python
import re

# Function to remove special characters (keep letters, numbers, and spaces)
def remove_special_characters(text):
    return re.sub(r"[^a-zA-Z0-9 ]", "", text)

# Apply to existing text columns
FinalJobs["job_role_and_duties"] = FinalJobs["job_role_and_duties"].astype(str).apply(remove_special_characters)
FinalJobs["requisite_skill"] = FinalJobs["requisite_skill"].astype(str).apply(remove_special_characters)
```
